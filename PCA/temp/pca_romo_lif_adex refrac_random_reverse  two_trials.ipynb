```python
import matplotlib.pyplot as plt
import numpy as np
import torch

from norse.torch.functional.lif_adex import LIFAdExParameters
from norse.torch.functional.lif_adex_refrac import LIFAdExRefracParameters

from tqdm import tqdm
from cgtasknet.net.lifadexrefrac import SNNlifadexrefrac
from cgtasknet.instruments.dynamic_generate import SNNStates
from cgtasknet.instruments.instrument_pca import PCA

from cgtasknet.net.states import LIFAdExRefracInitState
from cgtasknet.tasks.reduce import DefaultParams, RomoTask

dmparams1 = DefaultParams("RomoTask").generate_params()
dmparams1["delay"] = 0.1
dmparams1["trial_time"] = 0.1
dmparams1["values"] = (0.3, 0.8)
Task = RomoTask(dmparams1, mode="value")
dmparams2 = DefaultParams("RomoTask").generate_params()
dmparams2["delay"] = 0.1
dmparams2["trial_time"] = 0.1
dmparams2["values"] = (0.6, 0.9)
Task2 = RomoTask(dmparams2, mode="value")
dmparams3 = DefaultParams("RomoTask").generate_params()
dmparams3["delay"] = 0.1
dmparams3["trial_time"] = 0.1
dmparams3["values"] = (0.4, 0.1)
Task3 = RomoTask(dmparams3, mode="value")
dmparams4 = DefaultParams("RomoTask").generate_params()
dmparams4["delay"] = 0.1
dmparams4["trial_time"] = 0.1
dmparams4["values"] = (0.9, 0.7)
Task4 = RomoTask(dmparams4, mode="value")
feature_size = 2
output_size = 3
hidden_size = 250
batch_size = 1
neuron_parameters = LIFAdExRefracParameters(
    LIFAdExParameters(
        v_th=torch.as_tensor(0.65),
        tau_ada_inv = torch.as_tensor(1),
    ),
    rho_reset=torch.as_tensor(0.5),
)
model = SNNlifadexrefrac(
    feature_size,
    hidden_size,
    output_size,
    neuron_parameters=neuron_parameters,
    tau_filter_inv=500,
    save_states = True,
)
if True:
    model.load_state_dict(
        torch.load("..\models\\only_romo_lif_adex_refrac\\Only_romo_lif_adex_refrac_net_random_two_trials_250")
    )
init_state = LIFAdExRefracInitState(batch_size, hidden_size)
first_state = init_state.zero_state()
second_state = init_state.random_state()
inputs, target_out = Task.dataset(1, delay_between=200)
one_trajectory_time = inputs.shape[0] * 4 # - 2 + 1500 + 1500
v_mean = torch.zeros((one_trajectory_time,  hidden_size * 2 ))
number_of_trials = 10
#states_generator = SNNStates(model)
inputs1, target_out = Task.dataset(1, delay_between=200)
inputs1 = torch.from_numpy(inputs1).type(torch.float)

inputs2, target_out2 = Task2.dataset(1, delay_between=200)
inputs2 = torch.from_numpy(inputs2).type(torch.float)



inputs3, target_out3 = Task3.dataset(1, delay_between=200)
inputs3 = torch.from_numpy(inputs3).type(torch.float)

inputs4, target_out4 = Task4.dataset(1, delay_between=200)
inputs4 = torch.from_numpy(inputs4).type(torch.float)
data = np.concatenate((inputs1, inputs2, inputs3, inputs4))

for trial in tqdm(range(number_of_trials)):
    second_state = init_state.random_state()
    

    out1, states1 = model(inputs1, second_state)
    second_state = init_state.random_state()
    out2, states2 = model(inputs2, second_state)
    second_state = init_state.random_state()
    out3, states3 = model(inputs3, second_state)
    second_state = init_state.random_state()
    out4, states4 = model(inputs4, second_state)
    out = torch.concat((out1, out2, out3, out4))
    states = states1 + states2 + states3 + states4 
    v = []
    s = []
    i = []
    a = []
    for j in range(len(states)):
        v.append(states[j].lif_adex.v)
        s.append(states[j].lif_adex.z)
        i.append(states[j].lif_adex.i)
        a.append(states[j].lif_adex.a)

    s = torch.stack(s).detach()
    v = torch.stack(v).detach()
    a = torch.stack(a).detach()
    v_mean[:,  :hidden_size] += v[:, 0, :]
    v_mean[:, hidden_size:] += a[:, 0, :]
v_mean /= float(number_of_trials)
pca = PCA(3).decompose(v_mean.reshape(v_mean.shape[0], v_mean.shape[1]))
cmap = np.arange(0, len(v_mean))
data = np.concatenate((inputs1, inputs2, inputs3, inputs4))

%matplotlib widget
zeros_array = np.zeros((0, 1, 3))
plt.figure()
plt.plot(out[:, 0, :].detach().cpu().numpy()[:, 1])
plt.plot(out[:, 0, :].detach().cpu().numpy()[:, 2])

plt.show()

```
