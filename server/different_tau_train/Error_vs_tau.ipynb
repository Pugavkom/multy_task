```python
import matplotlib.pyplot as plt
import os
import matplotlib
plt.rcParams['font.size'] = 8
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
lw = 0.1
file = 'accuracy_dl_0.15'
x, y = [], []
with open(file, 'r') as f:
    line = f.readline()
    while(line):
        t_x, t_y = line.split(':')
        t_x = 1 / float(t_x)
        t_y = float(t_y) * 100
        x.append(t_x)
        y.append(t_y)
        line = f.readline()
plt.plot(x, y, '.', linestyle='--', linewidth=lw, label='$N=400$. With delay')     

print(y)
file = 'accuracy_without_dl_0.15'
x, y = [], []
with open(file, 'r') as f:
    line = f.readline()
    while(line):
        t_x, t_y = line.split(':')
        t_x = 1 / float(t_x)
        t_y = float(t_y) * 100
        x.append(t_x)
        y.append(t_y)
        line = f.readline()
plt.plot(x, y, '.', linestyle='--', linewidth=lw, label='$N=400$. Without delay')     
plt.xlim(min(x), 3)
plt.xscale("log", base=2)

ax.set_xticks(x, rotation=0)
print(x)
file = 'accuracy_dl_0.15_600'
x, y = [], []
with open(file, 'r') as f:
    line = f.readline()
    while(line):
        t_x, t_y = line.split(':')
        t_x = 1 / float(t_x)
        t_y = float(t_y) * 100
        x.append(t_x)
        y.append(t_y)
        line = f.readline()
plt.plot(x, y, '.', linestyle='--', linewidth=lw, label='$N=600$. With delay')  
file = 'accuracy_without_dl_0.15_600'
x, y = [], []
with open(file, 'r') as f:
    line = f.readline()
    while(line):
        t_x, t_y = line.split(':')
        t_x = 1 / float(t_x)
        t_y = float(t_y) * 100
        x.append(t_x)
        y.append(t_y)
        line = f.readline()
plt.plot(x, y, '.', linestyle='--', linewidth=lw, label='$N=600$. Without delay')  
plt.ylim([40, 100])
plt.grid()
plt.grid(which='minor', linestyle='--', alpha=.8)
plt.xlabel(r'$\tau_a$')
plt.ylabel(r'Accuracy, %')
ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
plt.legend()
plt.minorticks_on()
plt.tight_layout()

plt.savefig('/home/pugavko/Accuracy_vs_tau_a.eps')
```
